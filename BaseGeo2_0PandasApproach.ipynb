### Building the ordinal regression model

```python
import pymc as pm
import numpy as np
import pytensor.tensor as pt
import arviz as az
import matplotlib.pyplot as plt
```

```python

# from https://github.com/pymc-devs/pymc-examples/blob/main/examples/generalized_linear_models/GLM-ordinal-regression.ipynb
def constrainedUniform(N, min=0, max=1):
    return pm.Deterministic(
        "cutpoints",
        pt.concatenate(
            [
                np.ones(1) * min,
                pt.extra_ops.cumsum(pm.Dirichlet("cuts_unknown", a=np.ones(N - 2))) * (max - min)
                + min,
            ]
        ),
    )

K = 7 # number of choices on likert scale
num_groups = 5 # number of groups

def make_model(question, dataframe):

    with pm.Model() as model:

        # choice is made from https://github.com/pymc-devs/pymc-examples/blob/main/examples/generalized_linear_models/GLM-ordinal-regression.ipynb
        cutpoints = constrainedUniform(K, 0, K)

        # # use looser variant, also from pymc documentation
        # cutpoints_sigma = pm.Exponential("cutpoints_sigma", 1)
        # cutpoints = pm.Normal(
        #     "cutpoints",
        #     mu=np.linspace(0, K, K-1), # 0 indexed likert scale -> there should be K-1 cutpoints between 0 and K-1 
        #     sigma=cutpoints_sigma,
        #     transform=pm.distributions.transforms.ordered
        # )

        # # K=7 means we need exactly K-1 = 6 cutpoints
        # cutpoints_sigma = pm.Exponential("cutpoints_sigma", 1.0)

        # # Initialize cutpoints symmetrically around the middle of your scale (3.0)
        # cutpoints_init = np.linspace(1.0, 5.0, K - 1) 

        # cutpoints = pm.Normal(
        #     "cutpoints",
        #     mu=cutpoints_init,
        #     sigma=cutpoints_sigma,
        #     transform=pm.distributions.transforms.ordered, # Guarantees cutpoint[i] < cutpoint[i+1]
        #     initval=cutpoints_init
        # )



        slope_mu = pm.Normal("slope_mu", mu=0, sigma=1)
        slope_sigma = pm.HalfNormal("slope_sigma", sigma=1)
        
        # beta_time_groups = pm.Normal(
        #     "beta_time_groups",
        #     mu=mu_time_population,
        #     sigma=sigma_time_population,
        #     size=size
        # )

        # reparameterization trick
        slope_groups_offset = pm.Normal("slope_groups_offset", mu=0, sigma=1, size=num_groups)
        slope_groups = pm.Deterministic("slope_groups", slope_mu + slope_groups_offset * slope_sigma)

        intercept_mu_population = pm.Normal(
            "intercept_mu_population",
            mu=3, # middle of 0 indexed likert scale
            sigma=1,
        )
        intercept_sigma_population = pm.HalfNormal(
            "intercept_sigma_population", sigma=1
        )

        # beta_groups = pm.Normal(
        #     "beta_groups",  
        #     mu=mu_population,
        #     sigma=sigma_population,
        #     size=size
        # )

        # reparameterization trick
        intercept_groups_offset = pm.Normal("intercept_groups_offset", mu=0, sigma=1, size=num_groups)
        intercept_groups = pm.Deterministic("intercept_groups", intercept_mu_population + intercept_groups_offset * intercept_sigma_population)

        # combine group intercepts and group slopes to form the predictor
        # predictor = slope * time + intercept                     <- linear regression
        # prior for this predictor: no effect -> slope = 0, intercept = 3 (middle of 0 indexed likert scale)
        predictor = pm.Deterministic(
            "predictor", slope_groups[dataframe.institution] * dataframe.year + intercept_groups[dataframe.institution]
        )

        # likelihood
        y = pm.OrderedLogistic("y", cutpoints=cutpoints, eta=predictor, observed=dataframe[question].values.astype(np.int64))


        effect_after_7_years_groups = pm.Deterministic(
            "effect_after_7_years_groups", slope_groups * 7
        )
        effect_after_7_years_population = pm.Deterministic(
            "effect_after_7_years_population", slope_mu * 7
        )


        idata = pm.sample(
            nuts_sampler="numpyro",
            draws=2000,
            tune=2000,
            target_accept=0.995)
       
    return idata, model


def analyze(question, dataframe):
    # shift questions responses to be 0 indexed
    dataframe[question] = dataframe[question].astype(int) - 1

    # run mcmc
    idata, model = make_model(question, dataframe)

    return idata, model
    # az.summary(idata, var_names=["slope_mu", "slope_sigma", "intercept_mu_population", "intercept_sigma_population", "effect_after_7_years_population"])
    # az.plot_trace(idata, var_names=["slope_mu", "slope_sigma", "intercept_mu_population", "intercept_sigma_population", "effect_after_7_years_population"])
```

```python
# simulate a survey with distinct values for each group and year, and then fit the model to see if it can recover the parameters.

def make_data(shift):
    # np.random.seed(100)
    n_samples = 1000
    year = np.random.choice([0, 7], size=n_samples)
    institution = np.random.choice([0, 1, 2, 3, 4], size=n_samples)


    latent_rating = (
        0
        + (year == 0) * np.random.logistic(4.0, 0.5, size=n_samples)
        + (year == 7) * (institution == 0) * np.random.logistic(4.0-shift, 0.5, size=n_samples)
        + (year == 7) * (institution == 1) * np.random.logistic(4.0+shift, 0.5, size=n_samples)
        + (year == 7) * (institution == 2) * np.random.logistic(4.0, 0.5, size=n_samples)
        + (year == 7) * (institution == 3) * np.random.logistic(4.0, 0.5, size=n_samples)
        + (year == 7) * (institution == 4) * np.random.logistic(4.0, 0.5, size=n_samples)
    )


    # latent_rating = (
    #     4.0
    #     + shift * (year == 7) * (institution == 0)
    #     - shift * (year == 7) * (institution == 1)
    # ) + np.random.logistic(0, 0.5, size=n_samples)




    latent_rating = np.clip(latent_rating, 1, 7)
    explicit_rating = np.round(latent_rating)
    df = pd.DataFrame(
        {
            "latent_rating": latent_rating,
            "explicit_rating": explicit_rating,
            "year": year,
            "institution": institution
        }
    )
    return df

df = make_data(shift=0.25)
print(df.head())

# plot the distribution of explicit_rating for each institution and year
fig, ax = plt.subplots(2, 1, figsize=(10, 6))

bins = np.arange(-0.5, 9.5, 1)
ax[0].hist(df[df.year == 0][df.institution == 0].explicit_rating, bins=bins, histtype="step", label="0, uib", color="blue")
ax[0].hist(df[df.year == 0][df.institution == 1].explicit_rating, bins=bins, histtype="step", label="0, uibgeophys", color="green")
ax[0].hist(df[df.year == 0][df.institution == 2].explicit_rating, bins=bins, histtype="step", label="0, uio", color="orange")
ax[0].hist(df[df.year == 0][df.institution == 3].explicit_rating, bins=bins, histtype="step", label="0, uit", color="purple")
ax[0].hist(df[df.year == 0][df.institution == 4].explicit_rating, bins=bins, histtype="step", label="0, unis", color="red")
ax[1].hist(df[df.year == 7][df.institution == 0].explicit_rating, bins=bins, histtype="step", label="7, uib", color="blue")
ax[1].hist(df[df.year == 7][df.institution == 1].explicit_rating, bins=bins, histtype="step", label="7, uibgeophys", color="green")
ax[1].hist(df[df.year == 7][df.institution == 2].explicit_rating, bins=bins, histtype="step", label="7, uio", color="orange")
ax[1].hist(df[df.year == 7][df.institution == 3].explicit_rating, bins=bins, histtype="step", label="7, uit", color="purple")
ax[1].hist(df[df.year == 7][df.institution == 4].explicit_rating, bins=bins, histtype="step", label="7, unis", color="red")

for a in ax:
    a.set_xticks(np.arange(0, 9, 1))
    a.set_xlabel("explicit_rating")
    a.set_ylabel("count")
    a.legend()
```

```python
effects = {}
shifts = np.linspace(0, 1, 25)
for shift in shifts:
    df = make_data(shift=shift)
    print(f"{shift = }")
    idata, model = analyze("explicit_rating", df)
    
    # get the effects after 7 years for each group and the population
    effects[shift] = {
        "uib": idata.posterior['effect_after_7_years_groups'].sel(effect_after_7_years_groups_dim_0=0).values.flatten(),
        "uibgeophys": idata.posterior['effect_after_7_years_groups'].sel(effect_after_7_years_groups_dim_0=1).values.flatten(),
        "uio": idata.posterior['effect_after_7_years_groups'].sel(effect_after_7_years_groups_dim_0=2).values.flatten(),
        "uit": idata.posterior['effect_after_7_years_groups'].sel(effect_after_7_years_groups_dim_0=3).values.flatten(),
        "unis": idata.posterior['effect_after_7_years_groups'].sel(effect_after_7_years_groups_dim_0=4).values.flatten(),
        "population": idata.posterior['effect_after_7_years_population'].values.flatten()
    }

    # print(f"Shift: {shift:.2f}, Population effect after 7 years: mean = {np.mean(idata.posterior['effect_after_7_years_population'].values.flatten()):.3f}, std = {np.std(idata.posterior['effect_after_7_years_population'].values.flatten()):.3f}")
```
