## Run analysis

```python
from actors import Survey
from hierarchical_modeling import find_posterior_distribution
import arviz as az

from checks import check_pair_of_questions
from plots import sanitize_key
```

```python
def analyze_with_MCMC(questions_2019, questions_2026, question_text=None, group=None):

    # Check questions. 
    # it is not a gurantee that the prototype viable questions actually are viable
    # If there are problems, we do not perform the analysis
    status = check_pair_of_questions(questions_2019, questions_2026)
    if status != 0:
        # print(f"issue: question {question_text} had issues, skipping")
        return 0 #invalid question
    else:
        # print(f"check ok")
        # print(f"{question_text}")
        pass
    

    # after the sanity check, there might have been an axis correction, and we must therefore recount
    for q in questions_2019:
        q.counts = q.get_counts()
    for q in questions_2026:
        q.counts = q.get_counts()

    # final check
    for q in questions_2019 + questions_2026:
        if question_text != q.raw_text:
            return 0
        
    return 1 # <- if analysis already done, stop here

    trace_question1 = find_posterior_distribution(*questions_2019, quiet=0)
    trace_question2 = find_posterior_distribution(*questions_2026, quiet=0)
    
    scores_y1 = trace_question1.posterior['group_mean_scores'].values.reshape(-1, 5)
    scores_y2 = trace_question2.posterior['group_mean_scores'].values.reshape(-1, 5)
    population_mean_1 = trace_question1.posterior["population_mean_score"].values.flatten()
    population_mean_2 = trace_question2.posterior["population_mean_score"].values.flatten()


    np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_group_mean_score_2019.npy", scores_y1)
    np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_group_mean_score_2026.npy", scores_y2)
    np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_population_mean_score_2019.npy", population_mean_1)
    np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_population_mean_score_2026.npy", population_mean_2)

    # use arviz for a raw plot/diagnostics
    filename1 = f"Arviz raw posteriors {sanitize_key(questions_2019[0].raw_text)} 2019"
    filename2 = f"Arviz raw posteriors {sanitize_key(questions_2019[0].raw_text)} 2026"
    azplot1 = az.plot_trace_dist(trace_question1, combined=True)
    azplot2 = az.plot_trace_dist(trace_question2, combined=True)
    azplot1.add_title(filename1)
    azplot2.add_title(filename2)
    azplot1.savefig(f"..\\figures\\change_from_2019_to_2026\\Arviz plots\\{group}\\{filename1}.png")
    azplot2.savefig(f"..\\figures\\change_from_2019_to_2026\\Arviz plots\\{group}\\{filename2}.png")

    plt.close() # we dont need to see all the plots here

    print(az.summary(trace_question1, var_names=["frac", "conc"]))
    print(az.summary(trace_question2, var_names=["frac", "conc"]))

    return 1 # valid question
    
```

## Finding difference distributions

```python
# define the credible interval for the analysis
credible_interval = (2.5, 97.5) # <- 95% credible interval

# construct empty containers for the difference distributions. They are filled in the next two cells
difference_distributions_educators_dict = {
    "uib" : {},
    "uibgeophys" : {},
    "uio" : {},
    "uit" : {},
    "unis" : {},
    "population" : {}
}

credible_intervals_educators_dict = {
    "uib" : {},
    "uibgeophys" : {},
    "uio" : {},
    "uit" : {},
    "unis" : {},
    "population" : {}
}

difference_distributions_students_dict = {
    "uib" : {},
    "uibgeophys" : {},
    "uio" : {},
    "uit" : {},
    "unis" : {},
    "population" : {}
}

credible_intervals_students_dict = {
    "uib" : {},
    "uibgeophys" : {},
    "uio" : {},
    "uit" : {},
    "unis" : {},
    "population" :{}
}

# path of the results of the analysis
educator_scores_path = r"..\scores\change_from_2019_to_2026\educators"
student_scores_path = r"..\scores\change_from_2019_to_2026\students"
```

#### Educators

```python

for question_part_of_filename in list_of_questions_educators:
    
    # load the scores
    means_2019_groups_educators_filename = question_part_of_filename + "_group_mean_score_2019.npy"
    means_2026_groups_educators_filename = question_part_of_filename + "_group_mean_score_2026.npy"
    means_2019_population_educators_filename = question_part_of_filename + "_population_mean_score_2019.npy"
    means_2026_population_educators_filename = question_part_of_filename + "_population_mean_score_2026.npy"

    means_2019_groups_educators = np.load(educator_scores_path + f"\\{means_2019_groups_educators_filename}")
    means_2026_groups_educators = np.load(educator_scores_path + f"\\{means_2026_groups_educators_filename}")
    means_2019_population_educators = np.load(educator_scores_path + f"\\{means_2019_population_educators_filename}")
    means_2026_population_educators = np.load(educator_scores_path + f"\\{means_2026_population_educators_filename}")
    
    # find the difference distributions 
    difference_distribution_groups = means_2026_groups_educators - means_2019_groups_educators
    difference_distribution_population = means_2026_population_educators - means_2019_population_educators

    # split the groups
    difference_distribution_uib = difference_distribution_groups[:, 0]
    difference_distribution_uibgeophys = difference_distribution_groups[:, 1]
    difference_distribution_uio = difference_distribution_groups[:, 2]
    difference_distribution_uit = difference_distribution_groups[:, 3]
    difference_distribution_unis = difference_distribution_groups[:, 4]

    # find 95% credible intervals
    uib_lower_bound = np.percentile(difference_distribution_uib, credible_interval[0])
    uibgeophys_lower_bound = np.percentile(difference_distribution_uibgeophys, credible_interval[0])
    uio_lower_bound = np.percentile(difference_distribution_uio, credible_interval[0])
    uit_lower_bound = np.percentile(difference_distribution_uit, credible_interval[0])
    unis_lower_bound = np.percentile(difference_distribution_unis, credible_interval[0])
    population_lower_bound = np.percentile(difference_distribution_population, credible_interval[0])
    uib_upper_bound = np.percentile(difference_distribution_uib, credible_interval[1])
    uibgeophys_upper_bound = np.percentile(difference_distribution_uibgeophys, credible_interval[1])
    uio_upper_bound = np.percentile(difference_distribution_uio, credible_interval[1])
    uit_upper_bound = np.percentile(difference_distribution_uit, credible_interval[1])
    unis_upper_bound = np.percentile(difference_distribution_unis, credible_interval[1])
    population_upper_bound = np.percentile(difference_distribution_population, credible_interval[1])

    # add to the dictionary
    difference_distributions_educators_dict["uib"][question_part_of_filename] = difference_distribution_uib
    difference_distributions_educators_dict["uibgeophys"][question_part_of_filename] = difference_distribution_uibgeophys
    difference_distributions_educators_dict["uio"][question_part_of_filename] = difference_distribution_uio
    difference_distributions_educators_dict["uit"][question_part_of_filename] = difference_distribution_uit
    difference_distributions_educators_dict["unis"][question_part_of_filename] = difference_distribution_unis
    difference_distributions_educators_dict["population"][question_part_of_filename] = difference_distribution_population    

    credible_intervals_educators_dict["uib"][question_part_of_filename] = (uib_lower_bound, uib_upper_bound)
    credible_intervals_educators_dict["uibgeophys"][question_part_of_filename] = (uibgeophys_lower_bound, uibgeophys_upper_bound)
    credible_intervals_educators_dict["uio"][question_part_of_filename] = (uio_lower_bound, uio_upper_bound)
    credible_intervals_educators_dict["uit"][question_part_of_filename] = (uit_lower_bound, uit_upper_bound)
    credible_intervals_educators_dict["unis"][question_part_of_filename] = (unis_lower_bound, unis_upper_bound)
    credible_intervals_educators_dict["population"][question_part_of_filename] = (population_lower_bound, population_upper_bound)
    
```

```python

def plot_difference_distributions_curves(list_of_questions, difference_distributions_dict, credible_intervals_dict, group):
    return
    for question_part_of_filename in list_of_questions:
        
        # plot the difference dist for each group and for the total distribution
        # fig, axs=plt.subplots(2, 3, figsize=(12, 8), sharex=True, sharey=True)
        bins = 75
        vline_max = 0.2
        vline_min = -0.1
        # histograms
        # axs[0, 0].hist(difference_distributions_dict["uib"][question_part_of_filename], bins=bins, density=True)
        # axs[0, 1].hist(difference_distributions_dict["uibgeophys"][question_part_of_filename], bins=bins, density=True)
        # axs[0, 2].hist(difference_distributions_dict["uio"][question_part_of_filename], bins=bins, density=True)
        # axs[1, 0].hist(difference_distributions_dict["uit"][question_part_of_filename], bins=bins, density=True)
        # axs[1, 1].hist(difference_distributions_dict["unis"][question_part_of_filename], bins=bins, density=True)
        # axs[1, 2].hist(difference_distributions_dict["population"][question_part_of_filename], bins=bins, density=True)

        # plot as curves
        fig, axs=plt.subplots(2, 1, figsize=(8, 8), sharex=True, sharey=True)
        axs[0].hist(difference_distributions_dict["uib"][question_part_of_filename], histtype="step", bins=bins, density=True, label="UiB", color="blue")
        axs[0].hist(difference_distributions_dict["uibgeophys"][question_part_of_filename], histtype="step", bins=bins, density=True, label="UiB Geophysics", color="green")
        axs[0].hist(difference_distributions_dict["uio"][question_part_of_filename], histtype="step", bins=bins, density=True, label="UiO", color="red")
        axs[0].hist(difference_distributions_dict["uit"][question_part_of_filename], histtype="step", bins=bins, density=True, label="UiT", color="orange")
        axs[0].hist(difference_distributions_dict["unis"][question_part_of_filename], histtype="step", bins=bins, density=True, label="UNIS", color="purple")
        axs[1].hist(difference_distributions_dict["population"][question_part_of_filename], histtype="step", bins=bins, density=True, label="Population", color="black")
        
        # vlines
        axs[0].vlines([credible_intervals_dict["uib"][question_part_of_filename][0], credible_intervals_dict["uib"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="blue")
        axs[0].vlines([credible_intervals_dict["uibgeophys"][question_part_of_filename][0], credible_intervals_dict["uibgeophys"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="green")
        axs[0].vlines([credible_intervals_dict["uio"][question_part_of_filename][0], credible_intervals_dict["uio"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="red")
        axs[0].vlines([credible_intervals_dict["uit"][question_part_of_filename][0], credible_intervals_dict["uit"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="orange")
        axs[0].vlines([credible_intervals_dict["unis"][question_part_of_filename][0], credible_intervals_dict["unis"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="purple")
        axs[1].vlines([credible_intervals_dict["population"][question_part_of_filename][0], credible_intervals_dict["population"][question_part_of_filename][1]], vline_min, vline_max, label="95% cred. interval", color="black")

        # axs[0, 0].set_title("UiB", fontsize=ax_title_fs)
        # axs[0, 1].set_title("UiB Geophysics", fontsize=ax_title_fs)
        # axs[0, 2].set_title("UiO", fontsize=ax_title_fs)
        # axs[1, 0].set_title("UiT", fontsize=ax_title_fs)
        # axs[1, 1].set_title("UNIS", fontsize=ax_title_fs)
        axs[0].set_title("Groups", fontsize=ax_title_fs)
        axs[1].set_title("Population", fontsize=ax_title_fs)
        for ax in axs.flatten(): 
            ax.tick_params(axis="both", which="major", labelsize=fig_tick_fs, labelbottom=True)
            ax.legend()
            ax.set_xlim(-4,4)
            ax.hlines(0.0, -4, 4, color="black", linewidth=0.5)
        
        fig.supxlabel(r"$\mathbb{E}[X_\text{2026}] - \mathbb{E}[X_\text{2019}]$", fontsize=fig_axis_fs)
        fig.supylabel(r"Density", fontsize=fig_axis_fs)

        filename = f"{question_part_of_filename}"[:50]
        fig.suptitle(f"{question_part_of_filename}", fontsize=fig_title_fs)

        fig.tight_layout()
        print(f"saving as {filename}"+ " "*100, end="\r", flush=True)
        fig.savefig(f"..\\figures\\change_from_2019_to_2026\\difference_distributions\\{group}\\{filename}.png")

        plt.close()
```
