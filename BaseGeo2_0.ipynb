## Run analysis

```python
from actors import Survey
from hierarchical_modeling import find_posterior_distribution
import arviz as az

from checks import check_pair_of_questions
from plots import sanitize_key
```

```python
def analyze(questions_2019, questions_2026, question_text=None, load_from_previous=False, group=None):

    # Check questions. 
    # it is not a gurantee that the prototype viable questions actually are viable
    # If there are problems, we do not perform the analysis
    status = check_pair_of_questions(questions_2019, questions_2026)
    if status != 0:
        print(f"issue: question {question_text} had issues, skipping")
        return
    else:
        print(f"check ok")

    # after the sanity check, there might have been an axis correction, and we must therefore recount
    for q in questions_2019:
        q.counts = q.get_counts()
    for q in questions_2026:
        q.counts = q.get_counts()

    
    # if this analysis is already completed, we load it
    if load_from_previous:        
        print("loading...")
        # scores_y1 = np.load("..\\scores\\hierarchical_modeling\\score1.npy")
        # scores_y2 = np.load("..\\scores\\hierarchical_modeling\\score2.npy")
        # population_mean_1 = np.load("..\\scores\\hierarchical_modeling\\mean_pop_score1.npy")
        # population_mean_2 = np.load("..\\scores\\hierarchical_modeling\\mean_pop_score2.npy")

    else:
            
        trace_question1 = find_posterior_distribution(*questions_2019, quiet=0)
        trace_question2 = find_posterior_distribution(*questions_2026, quiet=0)
        
        scores_y1 = trace_question1.posterior['group_mean_scores'].values.reshape(-1, 5)
        scores_y2 = trace_question2.posterior['group_mean_scores'].values.reshape(-1, 5)
        population_mean_1 = trace_question1.posterior["population_mean_score"].values.flatten()
        population_mean_2 = trace_question2.posterior["population_mean_score"].values.flatten()
        np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_group_mean_score_2019.npy", scores_y1)
        np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_group_mean_score_2026.npy", scores_y2)
        np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_population_mean_score_2019.npy", population_mean_1)
        np.save(f"..\\scores\\change_from_2019_to_2026\\{group}\\{sanitize_key(questions_2019[0].raw_text)}_population_mean_score_2026.npy", population_mean_2)

        # use arviz for a raw plot/diagnostics
        filename1 = f"Arviz raw posteriors {sanitize_key(questions_2019[0].raw_text)} 2019"
        filename2 = f"Arviz raw posteriors {sanitize_key(questions_2019[0].raw_text)} 2026"
        azplot1 = az.plot_trace_dist(trace_question1, combined=True)
        azplot2 = az.plot_trace_dist(trace_question2, combined=True)
        azplot1.add_title(filename1)
        azplot2.add_title(filename2)
        azplot1.savefig(f"..\\figures\\change_from_2019_to_2026\\Arviz plots\\{group}\\{filename1}.png")
        azplot2.savefig(f"..\\figures\\change_from_2019_to_2026\\Arviz plots\\{group}\\{filename2}.png")
    


```
